```python
#número inverso: o número multiplicado pelo seu inverso é = 1. Inverso é núm * 1/núm = 1
#ex: 1 * 1/1 = 1; 2 * 1/2 = 1 (nesse caso, o inverso de 2 é 1/2).
```

O texto acima (número * inverso = neutro) escrito em notação matemática é:
$$x . x^{-1} = 1$$

---
Aplicando essa msm propriedade para matrizes:
$$ A . A^{-1} = I $$

Significa que **matriz . matriz inversa = matriz identidade**

**Equação fundamental da álgebra**
$$ A . X = Y $$
$$ A.A^{-1} . X = A^{-1}.Y $$
$$X = A^{-1}.Y$$

**linha 1**: matriz A vezes uma incógnita dá a matriz Y; **linha 2:** se multiplicarmos td por -1, temos essa saída - ñ entendi pq multiplicar por -1 e pq não multiplica o X tbm; **linha 3:** simplificação da equação considerando q A.A-1 é a I, que é neutra
(lembrar q essa solução só existe pq A é considerada **matriz quadrada**) A elevado a -1 é a matriz inversa.

Se A **não** fosse matriz quadrada na equação acima, precisaríamos criar uma matriz **pseudoinversa** pra colocar na equação.

$$ A . X = Y $$
$$ A.A^{-1} . X = A^{-1}.Y $$
$$X = A^{-1}.Y$$
$$X = (A^{T}.A)^{-1} . A^{T}.Y$$

-Essa última equação é usada no **Método dos Mínimos Quadrados** pra resolver problemas de regressão linear

-**pseudoinversa**: faz uma correção linear de A, com a matriz transposta, achando um valor mto próximo

-**A elevado a T é a matriz transposta**: inverter a linha e a coluna

```python
import matplotlib.pyplot as plt
```

**regressão linear:** vc tem os pontos e quer achar uma função (equação?) deles, tudo isso baseado em uma reta (por isso é linear)

```python
x1, y1 = -1, 0
x2, y2 = 0, 1
x3, y3 = 1, 2
x4, y4 = 2, 1

x = [x1, x2, x3, x4]
y = [y1, y2, y3, y4]

plt.figure(figsize=(10, 5))

plt.scatter(x1, y1, color = 'green', label="pares ordenados (x, y)")
plt.scatter(x2, y2, color = 'green')
plt.scatter(x3, y3, color = 'green')
plt.scatter(x4, y4, color = 'green')

plt.plot(x, y, label="gráfico que liga os pontos")
plt.grid(True)
plt.box(False)
plt.legend()
plt.title("Interpolação da reta sobre todos os pontos")
```

```python
#regressão linear acha 'm' e 'n' na seguinte equação, de forma a se ajustar melhor nos pontos acima.
#Equação: y = mx + n
#pra encontrar o valor de A, ir substituindo os X correspondentes na equação e achando o número q acompanha o m. N vai ser sempre 1.
#ex subs usando x1 e y1: 0 = m.(-1) + 1.n
A = np.array([[-1, 1], [0, 1], [1, 1], [2, 1]])
Y = np.array([[y1], [y2], [y3], [y4]])
```

```python
(m, n) = np.dot(np.dot(np.linalg.pinv(np.dot(A.T, A)), A.T), y)   #isso é a fórmula ali de cima em código: X=(AT.A)−1.AT.Y

print(m, n)
```

```python
x1, y1 = -1, 0
x2, y2 = 0, 1
x3, y3 = 1, 2
x4, y4 = 2, 1

x = [x1, x2, x3, x4]
y = [y1, y2, y3, y4]

plt.figure(figsize=(10, 5))

plt.scatter(x1, y1, color = 'green', label="pares ordenados (x, y)")
plt.scatter(x2, y2, color = 'green')
plt.scatter(x3, y3, color = 'green')
plt.scatter(x4, y4, color = 'green')

v_x = np.linspace(-1, 2)  #espaço em x que quero q a nova função com m e n atue (do -1 a 2, q está no gráfico)
plt.plot(x, y, label="gráfico que liga os pontos")
plt.plot(v_x, m*v_x + n, label="reta otimizada por MMQ - mínimos quadrados")
plt.grid(True)
plt.box(False)
plt.legend()
plt.title("Interpolação da reta sobre todos os pontos")
```

##Estatística

**Variável**: conceito - tipo de dado normalmente é imutável

**Tipos**

**Variável categórica**: ex: unidade federativa, faixas de idade (infância, adolescencia, etc.), quente e frio, com rótulos - sem ordem definida

**Variável discreta**: número/conjunto finito, possível contar, enumeráveis, números inteiros

**Variável contínua**: números reais, dízima, por ex: entre 6 e 7 há infinitos números, c/ tds as casas decimais q quiser

**Moda**: e qdo vc não tem repetição de números no conjunto? Se a frequência de todos os valores é igual? Nesse caso, é **amodal**, s/ moda. A moda é sempre o ponto mais alto do gráfico (pq o eixo x é qtde de valores)

**Distribuição normal**: média, mediana e moda coincidem no meio do gráfico, no ponto mais alto.

**Histograma**: distribuição de frequência

**Desvio padrão**: dispersão em relação à media.

**Z score** **= (x-méd)/std** é medida de padronização de modelo de dados (z score table)

```python
from sklearn.model_selection import train_test_split
```

```python
#Divide dados em variavel x e y. características são armazenadas na variável x.

x = df_diabetes.drop(["Class variable"], axis = 1)
x
```

```python
#Variável preditora/target (class variable, categoria) armazenar na variavel Y:

y = df_diabetes["Class variable"]
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
#modelo supervisionado: base que tem características e variável target (categoria):
#KNN: algoritmo de modelo sueprvisionado que classifica dados com base nos "vizinhos próximos": se várias mulheres com x características tem diabetes, posso prever se mulheres com caract parecidas tb tem.

from sklearn.neighbors import KNeighborsClassifier
```

```python
#passar a amostra de dados q o modelo vai usar pra treinar e pedir pra que treine (fit):

knn.fit(x_train, y_train)
```

##Testes de hipótese

```python
from scipy.stats import shapiro
```

```python
#primeiro, comprovar que a distribuição dos dados não é normal (é assimétrica), com teste de shapiro (teste paramétricos, teste P). Se der abaixo de 0.05, não é normal. Se não, é normal.

stat, p = shapiro(df_studio)
print("Soma do barulho dos álbuns de estúdio: ")
print("Estatística de teste: {:.4f}, valor p: {}".format(stat, p))

if p > 0.05:
  print("Não há evidência suficiente para rejeitar a hipótese de normalidade")
else:
  print("A hipótese de normalidade é rejeitada")
```

```python
stat, p = shapiro(df_ao_vivo)
print("Soma do barulho dos álbuns ao vivo: ")
print("Estatística de teste: {:.4f}, valor p: {}".format(stat, p))

if p > 0.05:
  print("Não há evidência suficiente para rejeitar a hipótese de normalidade")
else:
  print("A hipótese de normalidade é rejeitada")
```

```python
from scipy.stats import mannwhitneyu
```

```python
#teste pra comprovar se as duas distribuições (ao vivo e estúdio) são iguais ou não, teste U:

stat, p = mannwhitneyu(df_studio.sample(len(df_studio)), df_ao_vivo.sample(len(df_ao_vivo)), alternative="less")

print("Estatística de teste U: ", stat)
print("Valor p: ", p)

alpha = 0.05

if p < alpha:
  print("Diferença estatisticamente significante")
else:
  print("Não há diferença estatisticamente significante")
```

```python
#upsampling da classe minoritária (cria um volume igual à majoritária da minoritária em novo df)
df_minority_upsampled = resample(df_minority, replace=True, n_samples=len(df_majority), random_state=7)
```

```python
#juntando 2 dfs majoritário e minoritário em novo df:
df_equilibrado = pd.concat([df_majority, df_minority_upsampled])
```

```python
#ver q agora há msm número de dados com atraso vs sem atraso
sns.countplot(x="Class", data=df_equilibrado)
```

```python
x_equilibrado = df_equilibrado[["Flight", "Time", "Length", "DayOfWeek"]]
y_equilibrado = df_equilibrado["Class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_equilibrado, y_equilibrado, test_size=0.3, stratify=y_equilibrado, random_state=7)
```

```python
gnb_equilibrado = GaussianNB()
gnb_equilibrado.fit(x_train, y_train)
```

```python
y_pred_gnb_equilibrado = gnb_equilibrado.predict(x_test)
```
